```python
from urllib.parse import urlencode, urlparse, parse_qsl
import numpy as np
import pandas as pd
import requests
from sodapy import Socrata

# CDC API Keys and Token
from config import key_id, app_token, user_name, user_password
```

```python
# Example authenticated client (needed for non-public datasets):
client = Socrata("chronicdata.cdc.gov",
                  app_token,
                  username = user_name,
                  password= user_password)
# First 2000 results, returned as JSON from API / converted to Python list of
# dictionaries by sodapy.
results = client.get("g4ie-h725", limit=2000)

# Convert to pandas DataFrame
results_df = pd.DataFrame.from_records(results)

# 'https://soda.demo.socrata.com/resource/6yvf-kk3n.json?region=Virgin Islands region&source=pr'
```
